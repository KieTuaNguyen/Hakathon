## Part 3: Analysis

> Host with maximum listing

```python
# The hosts have the most listings on Airbnb
top_host = airbnb.host_name.value_counts().head(10)
top_host
```

```python
#setting figure size for future visualizations
sns.set(rc={'figure.figsize': (10, 8)})
sns.set_style('white')

top_host_df = pd.DataFrame(top_host)
top_host_df.reset_index(inplace=True)
top_host_df.rename(columns={'index': 'host_name',
                   'host_name': 'P_Count'}, inplace=True)
top_host_df
```

```python
viz_1 = sns.barplot(x="host_name", y="P_Count",
                    data=top_host_df, palette='dark:#5A9_r')
viz_1.set_title('Hosts with the most listings in BKK')
viz_1.set_ylabel('Count of listings')
viz_1.set_xlabel('Host Names')
viz_1.set_xticklabels(viz_1.get_xticklabels())
```

<ul>
  <li>As shown in the adjacent bar chart, we can see there is a good distribution among the top 10 hosts in Bangkok.</li>
  <li>The host named Noons has the second-highest number of listings at 91, the value of host named Noons is nearly half less than that of host named Curry.</li>
  <li>The host named K has the third highest number of listings of 86.</li>
</ul>

> Analysis of rental prices

```python
# Checking the descriptive statistics
airbnb_data_new = airbnb.drop(
    ['last_review', 'reviews_per_month'], axis=1)
airbnb_data_new.describe()
```

Looking at the statistics parameters of the features in the dataset. Average price is 117,42$ and it varies from 0$ to 900000$. 
<br>
900000$ seems to be very high price which can be a luxurious apartment.

```python
# Checking for outliers
airbnb_data_new.agg({'price':['mean','median','max','count']})
```

```python
#Plotting the boxplot of the price data.
plt.figure(figsize=(10, 5))
ax = sns.boxplot(y='price', data=airbnb_data_new).set_title(
    'Price Distribution by neighbourhood group')
sns.set_theme(style='white')
plt.xlabel('Neighbourhood')
plt.ylabel('Price')
plt.show()
```

As expected there are outliers present in the data. In order to fix this, we will be using the quantile based flooring and capping.
First, let's print the 10th and 90th percentile of the price column.

```python
# low quantile
q_low = airbnb_data_new['price'].quantile(0.10)
q_low
```

```python
#high quantile
q_high = airbnb_data_new['price'].quantile(0.9)
q_high
```

<ul>
  <li>We can see from the adjacent box plot that price data is skewed. There are outliers present in the data.</li>
  <li>We have eliminated the outliers based on quantile flooring and capping.</li>
  <li>We have considered price range which falls under the 10th percentile and 90th percentile range.</li>
  <li>The range is from 590$ to 3643$</li>
</ul>

After capping and flooring the price data, we get the below shown boxplot which is much cleaner.

```python
#removing the values below 10th percentile and above 90th percentile
airbnb_data_new = airbnb_data_new.drop(
    airbnb_data_new[airbnb_data_new['price'] < q_low].index)
airbnb_data_new = airbnb_data_new.drop(
    airbnb_data_new[airbnb_data_new['price'] > q_high].index)

#Create a new boxplot and check the result.
plt.figure(figsize=(10, 5))
ax = sns.boxplot(y='price', data=airbnb_data_new).set_title(
    'Price Distribution')
sns.set_theme(style='white')
plt.ylabel('Price')
plt.show()
```

> Busiest Host

```python
#Busiest hosts would be those who have maximum number of reviews as people are booking frequently at those hosts.
busiest_hosts = airbnb.groupby(['host_name', 'host_id', 'room_type',])[
    'number_of_reviews'].max().reset_index()
busiest_hosts = busiest_hosts.sort_values(
    by='number_of_reviews', ascending=False).head(10)
busiest_hosts
```

```python
# Visualising the busiest hosts data through bar plot.
host_name = busiest_hosts['host_name']
reviews = busiest_hosts['number_of_reviews']
plt.title('Busiest Hosts', {'fontsize': 18})
plt.xlabel('Host Names', {'fontsize': 18})
plt.ylabel('Number of reviews', {'fontsize': 18})
plt.bar(host_name, reviews)
```

<ul>
  <li>The adjacent bar plot shows the top 10 hosts with respect to number of reviews.</li>
  <li>Among them Suchada has highest numbers of reviews and we can assume that Suchada is the busiest host.</li>
  <li>The top hosts have listed private room, entire home/apt.</li>
</ul>

> Most Preferred Room Types

```python
# Plotting histogram for room types to look for which are most preferred room types.
plt.figure(figsize=(16, 8))
ax = sns.countplot(x='room_type', data=airbnb,
                   order=airbnb['room_type'].value_counts().index)
abs_values = airbnb['room_type'].value_counts(ascending=False)
rel_values = airbnb['room_type'].value_counts(
    ascending=False, normalize=True).values * 100
lbls = [f'{p[0]} ({p[1]:.0f}%)' for p in zip(abs_values, rel_values)]
# puts the labels on the bars
ax.bar_label(container=ax.containers[0], labels=lbls)
ax.set_title('Number of Listings by Room Type')
plt.xticks(rotation=0)
plt.show()
```

<ul>
  <li>Looking at the adjacent histogram, we can say that there are 4 room types listed in the entire dataset namely Entire home/apt, Private room, Shared room and Hotel room.</li>
  <li>Among this types the most preferred room type is Entire home/apt as well as Entire home/apt.</li>
  <li>Shared room is least preferred by people.</li>
</ul>

> Top 10 neighbourhoods with the most listing

```python
# Lets find the top 10 neighbourhoods with most listings
df1 = airbnb_data_new.groupby(['neighbourhood'])['id'].count().nlargest(10)
df1
```

```python
# Visualizing the top 10 neighbourhood listings through bar plot
plt.figure(figsize=(12, 6))
x = list(df1.index)
y = list(df1.values)
x.reverse()
y.reverse()

plt.title("Top 10 Neighbourhoods with the Most Listings", {'fontsize': 18})
plt.ylabel("Neighbourhood", {'fontsize': 18})
plt.xlabel("Total Listings", {'fontsize': 18})

plt.barh(x, y)
plt.show()
```

<ul>
  <li>From the various Neighbourhoods top 10 Neighbourhoods are listed in the chart.</li>
  <li>Khlong Toei has the highest numbers of listing which is around 1835.</li>
</ul>

> Grabbing top 10 neighbourhoods for sub-dataframe

```python
#grabbing top 10 neighbourhoods for sub-dataframe
sub_1 = airbnb.loc[airbnb['neighbourhood'].isin(['Khlong Toei', 'Vadhana', 'Huai Khwang', 'Sathon',
                                                                   'Phra Nakhon', 'Phra Khanong', 'Chatu Chak', 'Bang Na', 'Bang Rak', 'Ratchathewi'])]
#using catplot to represent multiple interesting attributes together and a count
viz_2 = sns.catplot(x='neighbourhood', hue='neighbourhood',
                    col='room_type', data=sub_1, kind='count')
viz_2.set_xticklabels(rotation=90)
```

> Scatter plot of room type

```python
#Scatter plot of room type
plt.figure(figsize=(10, 6))
sns.scatterplot(airbnb.longitude,
                airbnb.latitude, hue=airbnb.room_type)
plt.ioff()
```

> Using countplot to visualize the number of Airbnbs in each neighbourhood

```python
# using countplot to visualize the number of Airbnbs in each neighbourhood
plt.figure(figsize=(24, 8))
ax = sns.countplot(x='neighbourhood', data=airbnb,
                   order=airbnb['neighbourhood'].value_counts().index)
abs_values = airbnb['neighbourhood'].value_counts(
    ascending=False).values
ax.bar_label(container=ax.containers[0], labels=abs_values)

ax.set_title('Number of Listings by Neighbourhood')
plt.xticks(rotation=90)
plt.show()
```
